```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('zomato.csv',
usecols=['restaurant name','restaurant type','rate (out of 5)','num of ratings','avg cost (two people)','online_order','table booking','cuisines type','area','local address'])
```

#Data Cleaning


```python
df['rate (out of 5)'].fillna(3.4,inplace=True)
df['avg cost (two people)'].fillna(290,inplace=True)
```

```python
df.isnull().sum()
```

#Data Visualization

```python
df['area'].value_counts().head(10).plot(kind='bar',xlabel='Area',ylabel='Count',title=
'Top(10) Areas of Restaurants')
plt.xticks(rotation=85)
plt.show()
```

```python
df['rate (out of 5)'].plot(kind='hist')
plt.show()
```

```python
plt.hist(df['rate (out of 5)'], bins=5, edgecolor='black')
plt.xlabel('Rate (Out of 5)')
plt.ylabel('Frequency')
plt.title('Analysis of Rating')
plt.show()
```

```python
plt.scatter(df['rate (out of 5)'], df['avg cost (two people)'], alpha=1)
plt.xlabel('Rate (Out of 5)')
plt.ylabel('Average Cost (Two People)')
plt.title('Relation Between Rating and Cost')
plt.show()
```

```python
booking_counts = df['table booking'].value_counts()
labels = booking_counts.index
counts = booking_counts.values

plt.pie(counts, labels=labels,autopct='%1.1f%%', startangle=90)
plt.title('Table Booking Distribution')
plt.axis('equal')
plt.show()
```

```python
coun = df['online_order'].value_counts()
labels = coun.index
counts = coun.values

plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Online Order Distribution')
plt.axis('equal')
plt.show()
```
